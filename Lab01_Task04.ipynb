```python
# The Sorting Algorithm :

def Sorting(arr):
    for i in range(len(arr)):
        min= i
        for j in range(i+1,len(arr)):
            if arr[j][0] < arr[min][0]:
                min = j
            elif arr[j][0] == arr[min][0]:
                if arr[j][1] > arr[min][1]:
                    min = j
                elif arr[j][1] == arr[min][1]:
                    if arr[j][2] < arr[min][2]:
                        min = j
        arr[i], arr[min] = arr[min], arr[i]
    return arr
```

```python
# Task 04 :

text_cases = task4_input.readlines()
arr = []
num = int(text_cases[0].strip())
for i in range(1,num+1):
    train = text_cases[i].strip().split()
    n, t, d = train[0], train[-1], train[-3]
    arr.append([n,t,i,d])
sorted_train = Sorting(arr)
for i in range(len(sorted_train)):
    s = f'{sorted_train[i][0]} will departure for {sorted_train[i][3]} at {sorted_train[i][1]}'
    task4_output.write(f"{s}\n")
task4_output.close()
```
