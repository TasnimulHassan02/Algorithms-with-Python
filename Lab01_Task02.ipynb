```python
# Modified bubble sort algorithm:

def bubbleSort(arr):
    for i in range(len(arr)-1):
      swap = False
      for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap = True
      if not swap:
          break

# In the best-case scenario where the input array is already sorted, the inner loop will iterate only once without any swaps taking place. As a result, the outer loop will break terminating the sorting process. Thus, the time complexity of the code becomes Î¸(n) for the best-case scenario.
```

```python
# Task 02 :

for i in range(2):
  num_of_line = int(task2_input.readline().strip())
  line = task2_input.readline().strip().split(' ')
  arr = []

  for i in range(num_of_line):
    arr.append(int(line[i]))

  bubbleSort(arr)

  for i in arr:
    task2_output.write(str(i) + ' ')
  s = f'\n'
  task2_output.write(s)
```
