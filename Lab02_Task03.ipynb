```python
task3_input = open(file='/content/drive/MyDrive/CSE221/Lab 02/input3.txt', mode='r')
```

```python
task3_output = open(file='/content/drive/MyDrive/CSE221/Lab 02/output3.txt', mode='w')
```

```python
ln = int(task3_input.readline().strip())
arr = task3_input.readline().strip().split(' ')

def merge(left, right):
  merged = []
  i = 0
  j = 0
  while i < len(left) and j < len(right):
    if int(left[i]) < int(right[j]):
        merged.append(int(left[i]))
        i += 1
    else:
        merged.append(int(right[j]))
        j += 1
  while i < len(left):
    merged.append(int(left[i]))
    i += 1
  while j < len(right):
    merged.append(int(right[j]))
    j += 1
  return merged

def merge_sort(arr):
  if len(arr) <= 1:
    return arr
  mid = len(arr) // 2
  l = arr[:mid]
  r = arr[mid:]
  left = merge_sort(l)
  right = merge_sort(r)
  return merge(left, right)

sorted_list = merge_sort(arr)
task3_output.write(str(sorted_list))
```
