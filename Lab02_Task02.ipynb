```python
# 1. 0(nlogn) solution:----------
```

```python
l1 = int(task2_input.readline().strip())
list1 = task2_input.readline().strip().split(' ')
l2 = int(task2_input.readline().strip())
list2 = task2_input.readline().strip().split(' ')
final_list = list1 + list2

def merge(left, right):
  merged = []
  i = 0
  j = 0
  while i < len(left) and j < len(right):
    if int(left[i]) < int(right[j]):
        merged.append(int(left[i]))
        i += 1
    else:
        merged.append(int(right[j]))
        j += 1
  while i < len(left):
    merged.append(int(left[i]))
    i += 1
  while j < len(right):
    merged.append(int(right[j]))
    j += 1
  return merged

def merge_sort(arr):
  if len(arr) <= 1:
    return arr
  mid = len(arr) // 2
  l = arr[:mid]
  r = arr[mid:]
  left = merge_sort(l)
  right = merge_sort(r)
  return merge(left, right)

sorted_list = merge_sort(final_list)
task2_output.write(str(sorted_list))
```

```python
l1 = int(task2_input.readline().strip())
list1 = task2_input.readline().strip().split(' ')
l2 = int(task2_input.readline().strip())
list2 = task2_input.readline().strip().split(' ')
sorted_list = []
i = 0
j = 0
while i < l1 and j < l2:
  if int(list1[i]) < int(list2[j]):
    sorted_list.append(int(list1[i]))
    i += 1
  else:
    sorted_list.append(int(list2[j]))
    j += 1
while i < l1:
  sorted_list.append(int(list1[i]))
  i += 1
while j < l2:
  sorted_list.append(int(list2[j]))
  j += 1
task2_output.write(str(sorted_list))
```
