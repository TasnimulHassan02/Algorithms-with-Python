```python
task4_input = open(file='/content/drive/MyDrive/CSE221/Lab 02/input4.txt', mode='r')
```

```python
task4_output = open(file='/content/drive/MyDrive/CSE221/Lab 02/output4.txt', mode='w')
```

```python
ln = int(task4_input.readline().strip())
arr = task4_input.readline().strip().split(' ')

def find_max(arr):
  if len(arr) <= 1:
    return arr[0]
  mid = len(arr) // 2
  l = arr[:mid]
  r = arr[mid:]
  l_max = find_max(l)
  r_max = find_max(r)
  if int(l_max) > int(r_max):
    return l_max
  else:
    return r_max

s = find_max(arr)
task4_output.write(s)
```
