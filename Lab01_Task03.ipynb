```python
# Modified Selection sort algorithm :      (Performs the minimum number of swapping operations)

def SelectionSort(arr1, arr2):

  for i in range(len(arr1)):
    max_idx = i
    idx = i
    for j in range(i+1, len(arr1)):
        if arr2[max_idx] < arr2[j]:
            max_idx = j
            idx = j
        elif arr2[max_idx] == arr2[j] and arr1[max_idx] > arr1[j]:
          idx = j
    arr2[i], arr2[max_idx] = arr2[max_idx], arr2[i]
    arr1[i], arr1[idx] = arr1[idx], arr1[i]
```

```python
#Task 03 :

for i in range(2):
  num_of_std = int(task3_input.readline().strip())
  std_id = task3_input.readline().strip().split(' ')
  std_mark = task3_input.readline().strip().split(' ')
  id = []
  mark = []
  for i in range(num_of_std):
    id.append(int(std_id[i]))
    mark.append(int(std_mark[i]))

  SelectionSort(id, mark)

  for i in range(num_of_std):
    s = f'ID: {id[i]} Mark: {mark[i]}\n'
    task3_output.write(s)
  s = f'\n'
  task3_output.write(s)
```
