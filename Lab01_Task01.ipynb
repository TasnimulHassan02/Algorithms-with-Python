```python
# Task 01 (a) :

num_of_line = int(task1a_input.readline().strip())

for i in range(num_of_line):
    temp = int(task1a_input.readline().strip())
    if(temp%2==0):
        s = f"{temp} is an Even Number.\n"
        task1a_output.write(s)
    else:
        s = f"{temp} is a Odd Number.\n"
        task1a_output.write(s)
```

```python
task1b_input = open(file='/content/drive/MyDrive/CSE221/Lab 01/input1b.txt', mode='r')
```

```python
task1b_output = open(file='/content/drive/MyDrive/CSE221/Lab 01/output1b.txt', mode='w')
```

```python
# Task 01 (b) :

num_of_lines = int(task1b_input.readline().strip())

for i in range(num_of_lines):
  temp = task1b_input.readline().strip()
  temps = temp.split(' ')

  if temps[2] == '+':
    sum = int(temps[1]) + int(temps[3])
  elif temps[2] == '-':
    sum = int(temps[1]) - int(temps[3])
  elif temps[2] == '*':
    sum = int(temps[1]) * int(temps[3])
  elif temps[2] == '/':
    sum = int(temps[1]) / int(temps[3])

  s = f'The result of {temp.strip("calculate")} is {sum} \n'
  task1b_output.write(s)
```

```python
task1b_input.close()
```

```python
task1b_output.close()
```
